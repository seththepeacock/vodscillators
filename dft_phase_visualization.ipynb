```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set phase
phi = 0
f_basis = 11
f0_offset = 0.5
sr = 1000
t_max = 1
# Create sinusoid with sample rate 1000 and time 0 to 1 second
t = np.linspace(0, t_max, sr)
y = np.cos(2 * np.pi * (f_basis + f0_offset) * t + phi)
basis = np.cos(2 * np.pi * f_basis * t)

plt.plot(t, y, color="blue")
plt.plot(t, basis, color="red")
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fftshift

# Generate a test signal (e.g., a sinusoid)
fs = 1000  # Sampling frequency
t = np.arange(0, 1, 1/fs)  # Time vector
f = 51  # Frequency of the sinusoid
signal = np.sin(2 * np.pi * f * t)  # Signal

# Apply a Hann window centered on the signal
N = len(signal)
window = np.hanning(N)  # Create a Hann window of the same length as the signal
windowed_signal = signal * window  # Apply the window

# Plot the original signal and the windowed signal
plt.figure()
plt.plot(t, signal, label="Original Signal")
plt.plot(t, windowed_signal, label="Windowed Signal")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Signal with Hann Window Applied")
plt.show()

# signal = fftshift(signal)
# windowed_signal = fftshift(windowed_signal)

# Verify the phase shift (e.g., via FFT)
fft_original = np.fft.fft(signal)
fft_windowed = np.fft.fft(windowed_signal)

# Compute the phase of the original and windowed signals
phase_original = np.angle(fft_original)
phase_windowed = np.angle(fft_windowed)

# Plot the phase difference
plt.figure()
plt.scatter(np.fft.fftfreq(N, 1/fs), phase_original, label="Original Phase")
plt.scatter(np.fft.fftfreq(N, 1/fs), phase_windowed, label="Windowed Phase")
plt.legend()
plt.xlim(50, 52)
plt.xlabel("Frequency [Hz]")
plt.ylabel("Phase [rad]")
plt.title("Phase Comparison")
plt.show()
```
